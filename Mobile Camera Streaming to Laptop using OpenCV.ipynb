*Capture Video from mobile Camera on pc.*

Solution:

install DroidCamX on your mobile:

 https://play.google.com/store/apps/details?id=com.dev47apps.droidcamx

https://www.dev47apps.com/droidcam/linux/

and get the Browser IP Cam Accsess from app (e.g:http://192.168.1.3:4747) and add "/mjpegfeed" or "/video" and pass it to *cv.VideoCapture*

```python
import numpy as np
import cv2 as cv

# cap = cv.VideoCapture('http://192.168.1.3:4747/video')
cap = cv.VideoCapture('http://192.168.1.3:4747/mjpegfeed')

while cap.isOpened():
    ret, frame = cap.read()
    cv.imshow('frame', frame)
    if cv.waitKey(1) == ord('q'):
        break

cap.release()
cv.destroyAllWindows()
```

Or use program like DroidCam/DroidCamX or any other program that gets mobile camera on PC then read it using its index on your PC.

```python
import numpy as np
import cv2 as cv

cap = cv.VideoCapture(1)
if not cap.isOpened():
    print("Cannot open camera")
    exit()
    
while True:
    # Capture frame-by-frame
    ret, frame = cap.read()

    # if frame is read correctly ret is True
    if not ret:
        print("Can't receive frame (stream end?). Exiting ...")
        break
    # Display the resulting frame
    cv.imshow('frame', frame)
    if cv.waitKey(1) == ord('q'):
        break

# When everything done, release the capture
cap.release()
cv.destroyAllWindows()
```
